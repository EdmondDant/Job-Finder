2. huggingface pipeline

```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline

hf_pipe = pipeline(
    "text2text-generation",
    model="google/flan-t5-base",
    max_new_tokens=50,
    temperature=0.0,
    device=0  # use GPU
)

llm = HuggingFacePipeline(pipeline=hf_pipe)
```

```python
from langchain.prompts import PromptTemplate

prompt_template = PromptTemplate(
    input_variables=["user_input", "retrieved_examples"],
    template="""
You are a career advisor AI. Your job is to assign the best career sector based on a user's input.

Here are some labeled examples:
{retrieved_examples}

Now classify this:
"{user_input}"

Respond only in this format:
sector: <name of best-fit sector>""")

def combine_inputs(description, skill_set):
    if isinstance(skill_set, list):
        skills = ", ".join(skill_set)
    else:
        skills = skill_set
    return f"{description}\n\nRequired Skills: {skills}"
```

3. retrieval + chain

```python
from langchain.chains import LLMChain
from langchain.vectorstores import FAISS
from langchain.schema import Document
from langchain.embeddings import HuggingFaceEmbeddings
import re

def clean(text):
    return re.sub(r"\s+", " ", str(text).strip())

documents = [
    Document(
        page_content=clean(combine_inputs(row["job_description"], row["job_skill_set"])),
        metadata={"category": row["category"]}
    )
    for _, row in training.iterrows()
]

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = FAISS.from_documents(documents, embedding_model)
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

def make_examples(docs):
    return "\n".join([
        f'"{doc.page_content[:300]}" → sector: {doc.metadata["category"]}'
        for doc in docs
    ])


```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")

model.save_pretrained("career_model")
tokenizer.save_pretrained("career_model")
```

```python
chain = LLMChain(llm=llm, prompt=prompt_template)

def classify_sector(query):
    top_docs = retriever.invoke(query)
    print("Top retrieved examples:\n")
    for doc in top_docs:
        print(f'"{doc.page_content[:300]}" → sector: {doc.metadata["category"]}')
        print("---")

    examples = make_examples(top_docs)
    response = chain.invoke({"user_input": query, "retrieved_examples": examples})
    return response["text"].strip()

```

```python
print(classify_sector("I want to serve my country through discipline and strength in defense"))
# → sector: Military
```
